# Beautifulsoup module for HTML scraping

```python
from bs4 import BeautifulSoup
import sys
```

```python
fh = open("test.html")
soup = BeautifulSoup(fh)
```

```python
table = soup.find(id="svcnv_subset").table
```

```python
headers = [header.text for header in table.find_all('th')]
```

```python
headers
```

```python
parseRow = lambda r: [ cell.text.encode('utf8') for cell in r.find_all('td')]
```

```python
rows = [parseRow(row) for row in table.find_all('tr')]
```
